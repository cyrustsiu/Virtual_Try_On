```python
from Model import Model
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import time
import cv2

import os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

%matplotlib inline
```

```python
model = Model("checkpoints/jpp.pb",
              "checkpoints/gmm.pth", 
              "checkpoints/tom.pth")
```

```python
img = np.array(Image.open("../Virtual-Try-On-Flask/data/test/image/000066_0.jpg"))
plt.imshow(img)
```

```python
c_img = np.array(Image.open("example/cloth.jpg"))
plt.imshow(c_img)
```

```python
start = time.time()
result,trusts = model.predict(img, c_img, need_pre=False,check_dirty=True)
if result is not None:
    end = time.time()
    print("time:"+str(end-start))
    print("Confidence"+str(trusts))
    plt.imshow(result)
```
